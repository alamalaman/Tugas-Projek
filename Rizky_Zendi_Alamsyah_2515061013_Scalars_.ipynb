### Scalars (Rank 0 Tensors) in Base Python

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x = 13
x
```

```python
# tampilkan tipe data dari x
type(x)
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y = 29
```

```python
# tampilkan tipe data dari y
type(y)
```

```python
py_sum = x + y
py_sum
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = 6.0
```

```python
# jumlahkan x_float dengan y
float_sum = x_float + y
float_sum
```

```python
# tampilkan tipe data dari x_float
type(float_sum)
```

### Scalars in TensorFlow (version 2.0 or later)

Tensors created with a wrapper, all of which [you can read about here](https://www.tensorflow.org/guide/tensor):  

* `tf.Variable`
* `tf.constant`
* `tf.placeholder`
* `tf.SparseTensor`

Most widely-used is `tf.Variable`, which we'll use here.

As with TF tensors, in PyTorch we can similarly perform operations, and we can easily convert to and from NumPy arrays.

Also, a full list of tensor data types is available [here](https://www.tensorflow.org/api_docs/python/tf/dtypes/DType).

```python
import tensorflow as tf
```

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x_tf = tf.Variable(13, dtype=tf.int16)
x_tf
```

```python
x_tf.shape
```

```python
y_tf = tf.Variable(29, dtype=tf.int16)
```

```python
x_tf + y_tf
```

```python
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
tf_sum.numpy() # note that NumPy operations automatically convert tensors to NumPy arrays, and vice versa
```

```python
type(tf_sum.numpy())
```

```python
tf_float = tf.Variable(25., dtype=tf.float16)
tf_float
```
